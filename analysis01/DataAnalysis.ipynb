#### 간단한 웹사이트 크롤링 하기 (뉴스페이지 - 언론사별 가장 많이 본 뉴스)
<br>

```python
# 2.✔ 원하는 사이트로 이동하는 코드 

from selenium import webdriver

driver = webdriver.Chrome('./chromedriver')  #chorme 브라우저 띄움

try:
    driver.get('https://news.naver.com/')   #naver news 페이지로 이동
    elem = driver.find_element_by_class_name('_officeTopRanking1087472')  #해당하는 class값 찾음
    lis = elem.find_elements_by_tag_name('li')  #elem 클래스 안에 있는 태그 중 li 태그 찾음
    
    for li in lis:  #해당 li의 text값 모두 출력
        divtag = li.find_element_by_class_name('list_text_inner')
        atag = divtag.find_element_by_class_name('list_tit')
        print(atag.text)
        print(atag.get_attribute('href'))
        
    input()  #대기
    
except Exception as e:
    print(e)

finally:
    driver.quit()
    
```

#### 간단한 웹사이트 수집하기 (블로그 - 제목과 링크)
-  xpath('./ul/li') 현재 ./태그에 있는 ul>li를 불러올 수 있다
<br>

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys 

driver = webdriver.Chrome('./chromedriver')

try:
    driver.get('https://www.naver.com/')
    elem = driver.find_element_by_id('query')
    elem.send_keys('취업')  # elem태그에 키 값 패스트캠퍼스를 전달
    elem.send_keys(Keys.RETURN)  #RETURN은 키보드 enter를 의미
    
    elem = driver.find_element_by_class_name('list_news')  #해당 클래스 가지고 있는 태그 가져옴
    lis = elem.find_elements_by_class_name('bx')
    
    for li in lis:
#         print(li.text)
        atag = li.find_element_by_class_name('news_tit') 
        title =atag.get_attribute('title')
        if not title:   #title 속성값 없는 경우
            title=atag.text
        print(atag.get_attribute('title'))  #title의 속성값을 가져온다
        print(atag.get_attribute('href'))
    
    input()
    
except Exception as e:
    print(e)
finally:
    driver.quit()
```

#### Gmail 이용해서 메일 보내기
<br>

```python
from email.mime.text import MIMEText
from email.mime.multipart import MIMEMultipart
import smtplib

#SMTP 서버에 접속할 수 있는 정보

SMTP_SERVER = 'smtp.gmail.com'
SMTP_PORT = 465
SMTP_USER = '<보내는 사람 이메일>@gmail.com'
SMTP_PASSWORD = "ulykiboanczbtrgz"

#이메일 발송

def send_mail(name, addr, subject, contents):
    msg = MIMEMultipart('alternative')
    msg['From'] = SMTP_USER
    msg['To'] = addr
    msg['Subject'] = name + '님, '+ subject
    
    text = MIMEText(contents, _charset='utf-8')
    msg.attach(text)
    
    smtp = smtplib.SMTP_SSL(SMTP_SERVER, SMTP_PORT)
    smtp.login(SMTP_USER,SMTP_PASSWORD)
    smtp.sendmail(SMTP_USER, addr, msg.as_string())
    smtp.close()

contents='''안녕하세요
자동화 메일입니다
'''
    
send_mail('손희정','<받는사람 이메일>@naver.com','자동화 메일',contents)
```
