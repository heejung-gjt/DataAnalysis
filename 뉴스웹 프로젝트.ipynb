```python
# 인사이트 컨텐츠 정치 전체적인 html 크롤링

import requests
from bs4 import BeautifulSoup

for num in range(1,6):
    response = requests.get(f'https://www.insight.co.kr/section/politics?page={num}')
    soup = BeautifulSoup(response.text, 'html.parser')

    # 메인 페이지에 있는 뉴스 ul문
    ul = soup.select_one('body > div.content > div > div.section-wrap > div.section-list > ul')
    # 메인 페이지에 있는 뉴스 리스트들
    lis = ul.select('li')

    # 뉴스기사 list for문
    for li in lis:
        #뉴스기사 세부 내용 들어갈 수 있는 a링크 가져오기     
        url = li.select_one('div > a.section-list-article-title')['href']

        #뉴스기사 상세페이지 접속, 다시한번 request,beautifulsoup써서 html parser한 후 해당되는 html가져오기
        #raw_each는 뉴스기사 상세페이지 url, html_each는 상세페이지 html.parser진행     
        raw_each = requests.get(url)
        html_each = BeautifulSoup(raw_each.text, 'html.parser')
        #select_one써서 title html형태 가져오기(.text찍으면 컨텐츠만 가져올 수 있음)     
        print(html_each.select_one('body > div.content > div > div.news-container > div.news-header'))
        #select_one써서 content,img부분 html형태로 가져오기(.text찍으면 img는 가져올 수 없음 img는['src'])     
        print(html_each.select_one('body > div.content > div > div.news-container > div.news-wrap > div.news-article > div.news-article-memo'))
        print()
        print()
```

```python
# 인사이트 카테고리, 프리뷰 이미지src, title, 날짜 

import requests
from bs4 import BeautifulSoup

for num in range(1,6):
    response = requests.get(f'https://www.insight.co.kr/section/money?page={num}')
    soup = BeautifulSoup(response.text, 'html.parser')

    # 메인 페이지 기사 ul
    ul = soup.select_one('body > div.content > div > div.section-wrap > div.section-list > ul')
    # 메인 페이지 기사리스트들 li
    lis = ul.select('li')
    # 현재 카테고리 네임
    new_category = soup.select_one('body > div.content > div > div.section-wrap > div.section-gnb > span')

    # 기사 리스트 for문
    # 프리뷰 기사 img, title, date가져옴
    # 현재 카테고리 각각의 리스트에 함께 나오게 하기 위해 단순히 위에서 받은 new_category밑에다 써줌
    # if문 적은 이유는 기사이름이 날짜와 함께 나오는 경우가 있음 "." 이 있는 경우는 기사 네임이 있는 경우라 .을 기준으로 나누어서 날짜만 가져옴
    for li in lis:
        preview_img = li.select_one('li > div > a.section-list-article-image')
        preview = li.select_one('li > div')
        preview_img = preview_img['style'].split("(")[1].split(')')[0]
        preview_title = li.select_one('li > div > a.section-list-article-title')
        preview_date = li.select_one('li > div > span.section-list-article-byline')
        print(new_category.text.split('>')[1].strip())
        print(preview_img)
        print(preview_title.text.strip())
        if '·' in preview_date.text:
            print(preview_date.text.strip().split('·')[1])
        else:
            print(preview_date.text.strip())
        print()
```
